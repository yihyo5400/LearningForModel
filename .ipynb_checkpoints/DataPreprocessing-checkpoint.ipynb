```python
# Date 칼럼에서 년도, 월, 일 정보를 숫자 형태로 반환하는 함수를 작성합니다.
def date_encoder(date):
    day, month, year = map(int, date.split('/'))
    return day, month, year

# 이 함수를 Date 칼럼에 적용하여 Day, Month, Year 칼럼을 생성합니다.
train['Day'] = train['Date'].apply(lambda x: date_encoder(x)[0])
train['Month'] = train['Date'].apply(lambda x: date_encoder(x)[1])
train['Year'] = train['Date'].apply(lambda x: date_encoder(x)[2])

# 결과를 확인합니다.
train
```

```python
target = train['Weekly_Sales']
```

```python
# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=2022)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
X_train = X_train.drop(columns=['Weekly_Sales'])
```

```python
X_val = X_val.drop(columns=['Weekly_Sales'])
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.metrics import mean_squared_error

def rmse(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
```

```python
# 랜덤포레스트
from sklearn.ensemble import RandomForestRegressor #RandomForestClassifier는 범주형종속변수에만 사용!
model = RandomForestRegressor(random_state=2022)
model.fit(X_train, y_train)
pred = model.predict(X_val)
print(rmse(y_val, pred))
```

```python
# 서포트벡터머신 - 비선형
from sklearn.svm import SVR
model = SVR()
model.fit(X_train, y_train)
pred = model.predict(X_val)
print(rmse(y_val, pred))
```

```python
# 서포트벡터머신 - 선형
from sklearn.svm import LinearSVR
model = LinearSVR(random_state=2022)
model.fit(X_train, y_train)
pred = model.predict(X_val)
print(rmse(y_val, pred))
```
